### Importing The Libraries And Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart.csv')
```

### Exploring the Dataset

```python
data.head()
```

- age
- sex
- chest pain type (4 values)
 - Value 0: typical angina
 - Value 1: atypical angina
 - Value 2: non-anginal pain
 - Value 3: asymptomatic 
-  trestbps: resting blood pressure (in mm Hg on admission to the hospital) 
- chol: serum cholestoral in mg/dl 
- fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 
-  restecg: resting electrocardiographic results
 - Value 0: normal
 - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
 - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria 
- thalach: maximum heart rate achieved 
- exang: exercise induced angina (1 = yes; 0 = no) 
- oldpeak = ST depression induced by exercise relative to rest 
-  slope: the slope of the peak exercise ST segment
 - Value 1: upsloping
 - Value 2: flat
 - Value 3: downsloping
- ca: number of major vessels (0-3) colored by flourosopy 
- thal: 3 = normal; 6 = fixed defect; 7 = reversable defect 
- target : 0=less chance of heart attack, 1=more chance of heart attack

### How Many People Have Heart Disease, And How Many Don't Have Heart Disease In This Dataset ?

```python
data['target'].value_counts()
```

```python
cp1 = sns.countplot( x = 'target', data = data)
for bars in cp1.containers : cp1.bar_label(bars)
plt.xticks([0,1],['No Disease','Disease'])
plt.show
```

### Which Sex Has The Most Heart Disease [ Male or Female ]?

```python
data['sex'].value_counts()
```

```python
cp2 = sns.countplot( x = 'sex', data = data )
plt.xticks([0,1],['Female','Male'])
for bars in cp2.containers : cp2.bar_label(bars)
```

###  Find Gender Distribution According to The Target Variable.

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

###  Check Age Distribution In The Dataset

###  Check Chest Pain Type

- chest pain type (4 values)
 - Value 0: typical angina
 - Value 1: atypical angina
 - Value 2: non-anginal pain
 - Value 3: asymptomatic 

```python
sns.countplot(data = data, x = 'cp')
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal pain","asymptomatic "])
plt.xticks(rotation=7)
plt.show()
```

###  Show The Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal pain","asymptomatic "])
plt.xticks(rotation = 5)
plt.show()
```

### Show Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

###  Compare Resting Blood Pressure As Per Sex Column

```python
g = sns.FacetGrid(data,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',fill=True)
plt.legend(labels=['Male','Female'])
```
